```python
# load image

import matplotlib.pyplot as plt
img = plt.imread('peppers.jpg'
plt.imshow(img) #show image
plt.axis('off')
plt.show()
```

```python
img_flat = img.reshape(img.shape[0]*img.shape[1],3)
print(img_flat.shape)
img_flat[:5,:]
img_flat.shape
```

```python
# Apply k-means Algorithm

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(img_flat)
kmeans.labels_
```

```python
# loops for each cluster center
# replace kmeans.labels_ with kmeans.cluster_centers_[i]

import numpy as np
img_flat_2 = img_flat.copy()
for i in np.unique(kmeans.labels_):
    img_flat_2[kmeans.labels_==i,:] = kmeans.cluster_centers_[i]
```

```python
img2 = img_flat_2.reshape(img.shape)
plt.imshow(img2)
plt.axis('off')
plt.show()
```

#Question 3 Apply elbow method

```python
all_inertia=[]
for i in range (1,8):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(img_flat)
    all_inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1,8),all_inertia)
plt.show()
```
